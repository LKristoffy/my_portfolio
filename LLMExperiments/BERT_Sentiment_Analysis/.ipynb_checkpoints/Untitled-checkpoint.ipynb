```python
# Plotting the distribution for dataset.
ax = train.groupby('label').count().plot(kind='bar', title= 'Distribution of data', legend=False)
ax.set_xticklabels(['Positive', 'Negative'], rotation=0)
# Storing data in list
text, sentiment = list(train['tweet']), list(train['label'])
```

```python
labels = train['label'].tolist()
tweets = train['tweet'].tolist()
```

```python
print(tweets[10])

tokeniser = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

inputs = tokeniser(tweets, padding='max_length', truncation = True)
```

```python
class twitterDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
        
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
    
    def __len__(self):
        return len(self.labels)
    
    
train_dataset = twitterDataset(inputs, labels)
```

```python
print(train_dataset.__getitem__(10))
```

```python
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)
trainer.train()
```

```python
#Load model
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)
model.load_state_dict(torch.load("/Users/liamkristoffy/Documents/Code Projects/Sentiment Analysis/BERTModel.pt"))
model.eval()
```

```python
#Function to evaluate tweet without training model
def to_check_results(test_encoding):
    input_ids = torch.tensor(test_encoding["input_ids"]).to(device)
    attention_mask = torch.tensor(test_encoding["attention_mask"]).to(device)
    with torch.no_grad():
        outputs = model(input_ids.unsqueeze(0), attention_mask.unsqueeze(0))
        
    y=np.argmax(outputs[0].to('cpu').numpy())
    
    return y
```

```python
l2 = []
#Find model prediction for all test tweets
for i in test["tweet"]:
    test_encoding1 = tokeniser(i, truncation=True, padding=True)
    input_ids = torch.tensor(test_encoding1["input_ids"]).to(device)
    attentinos_mask = torch.tensor(test_encoding1["attention_mask"]).to(device)
    op = to_check_results(test_encoding1)
    l2.append(op)
```
