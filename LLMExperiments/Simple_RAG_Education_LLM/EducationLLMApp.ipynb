```python
all_paper_paths = [f for f in os.listdir('/content/pdf_papers') if f.endswith(".pdf")]
print(all_paper_paths)
```

```python
docs = []
for path in all_paper_paths:
  loader = PyPDFLoader(os.path.join('/content/pdf_papers',path))
  data = loader.load()
  splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
  docs.extend(splitter.split_documents(data))

```

```python
print(docs)
```

```python

for doc in docs:
  print(doc.metadata)
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
import os

# Embed the documents in a persistent Chroma vector database
embedding_function = OpenAIEmbeddings(openai_api_key=openai_api_key)
vectorstore = Chroma.from_documents(
    docs,
    embedding=embedding_function,
    persist_directory= os.path.join(os.getcwd(), 'vector_store2')
)

```

```python
from langchain_core.prompts.chat import ChatPromptTemplate

# Add placeholders to the message string
message = """
You are a study guide for students studying GCSE maths. Please provide 3 question and answer pairs as practice material for the topic of their choosing in GCSE maths, with the context provided.
The answer should be clear, taking things logically step-by-step, and be correct given the provided question you design.

Context:
{context}

Topic:
{topic}

Question/ Answer pairs:
"""

# Create a chat prompt template from the message string
prompt_template = ChatPromptTemplate.from_messages([("human", message)])
```

```python
# Configure the vector store as a retriever
retriever = vectorstore.as_retriever(
    search_type="similarity",
    search_kwargs={'k':6}
)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnableParallel, RunnablePassthrough

llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0, openai_api_key=openai_api_key)

# Create a chain to link retriever, prompt_template, and llm
rag_chain = ({"context": retriever, "topic": RunnablePassthrough()}
            | prompt_template
            | llm)

# Invoke the chain
response = rag_chain.invoke("geometry")
print(response.content)
```
